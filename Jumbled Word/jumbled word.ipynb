# Jumbled Word

```python
import random
```

## Getting Data 

```python
word_list = open("MyWord.txt").readlines()
```

## Data Cleaning

```python
word_list = list(map(str.strip, word_list))
word_list = [word for word in word_list if word.isalpha()]
word_list = list(map(str.lower, word_list))
```

**Code for jumbled word game.**

```python
print("****Welcom to the Word Jumble****\n")
p1 = input("Enter player1 name: \n")
p2 = input("Enter player2 name: \n")
score_p1 = 0
score_p2 = 0
while True:
    words = list(map(str.lower, word_list))
    pick = random.choice(words)
    pick = pick.lower()
    word = random.sample(pick, len(pick))
    jumbled = ''.join(word)
    print(p1+ " your turn..\n") 
    print(jumbled)
    w = input("\nWhat do you think?\n")
    if w == pick:
        print('\nCORRECT')
        print(p1+' you got 1 point\n')
        score_p1 += 1
    else:
        print('\nWrong Answer')
        print('Better luck next time\n')
        print(p2+" your turn..\n")   
        w = input("\nWhat do you think?\n")
        if w == pick:
            print('\nCORRECT')
            print(p2+' you got 1 point\n')
            score_p2 += 1
        else:
            print('\nWrong Answer')
            print("better luck next time\n")
            print("correct word is\n", pick)
            q = input("DO you want to play again? y/n\n")
            if q == 'n':
                break
            else:
                continue
        pick = random.choice(words)
        word = random.sample(pick, len(pick))
        jumbled = ''.join(word)
        print(p1+ " your turn..\n") 
        print(jumbled)
        w = input("\nWhat do you think?\n")
        if w == pick:
            print('\nCORRECT')
            print(p1+' you got 1 point\n')
            score_p1 += 1
        else:
            print('\nWrong Answer')
            print('Better luck next time\n')
            print(p2+" your turn..\n")
            w = input("\nWhat do you think?\n")
            if w == pick:
                print('\nCORRECT')
                print(p2+' you got 1 point\n')
                score_p2 += 1
            else:
                print('\nWrong Answer')
                print("better luck next time\n")
                print("correct word is\n", pick)
                q = input("DO you want to play again? y/n\n")
                if q == 'n':
                    break
                else:
                    continue
                    
    print(p2+" your turn..\n")
    pick = random.choice(words)
    word = random.sample(pick, len(pick))
    jumbled = ''.join(word)
    print(jumbled)
    w = input("\nWhat do you think?\n")
    if w == pick:
        print('\nCORRECT')
        print(p2+ ' you got 1 point\n')
        score_p2 += 1
    else:
        print('\nWrong Answer')
        print('Better luck next time\n')
        print(p1+" your turn..\n")
        w = input("\nWhat do you think?\n")
        if w == pick:
            print('\nCORRECT')
            print(p1+ ' you got 1 point\n')
            score_p1 += 1
        else:
            print('\nWrong Answer')
            print('better luck next time\n')
            print("correct word is\n", pick)
            q = input("Do you want to play again? y/n \n")
            if q == 'n':
                break
            else:
                continue
    q = input("Do you want to play more? y/n \n")
    if q == 'n':
        break
    else:
        continue
print(p1+ " your score is \n",score_p1)
print(p2+ ' your score is \n',score_p2)
if score_p1>score_p2:
    print("Congratulations " +p1+ " you won\n")
elif score_p1<score_p2:
    print("Congratulations " +p2+ ' you won\n')
else:
    print("It's a tie\n")
print("Thanks for playing")
```
